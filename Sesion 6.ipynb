# Visualizacion de datos usando python

```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

```python
df.columns
```

## Variables Categoricas

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil-' + str(int(x)))
```

### Grafica de barras

```python
figuras, axis = plt.subplots() 
axis.bar(y.value_counts().index, y.value_counts())
axis.set_xlabel('Cuartiles del a√±o')
axis.set_ylabel('Frecuencias')
```

### Usando colores

```python
figuras, axis = plt.subplots() 
axis.bar(y.value_counts().index, y.value_counts())
axis.set_xlabel('Cuartiles del a√±o')
axis.set_ylabel('Frecuencias')
axis.patches[0].set_facecolor('red')
```

### Grafica circular

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
ax.set_title('Diagrama de pie')
```

## Variables Numericas

```python
y = df['viento']
```

```python
fig, ax= plt.subplots()
ax.hist(y, bins = 30)
#bins son las partes en que se divide el histogram
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencias')
plt.axvline(np.mean(y), c ='r', linestyle = '--', label = 'Promedio')
plt.axvline(np.mean(y)+np.std(y), c ='k', linestyle = '--', label = '+ 1-desviacion')
plt.axvline(np.mean(y)-np.std(y), c ='k', linestyle = '--', label = '- 1-desviacion')
ax.legend()
#es necesaio definir los labels para poder aplicar el legend
```
